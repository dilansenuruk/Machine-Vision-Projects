```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# defining the datapoints for the analysis
X = np.array([[ 99,  -1], [ 98,  -1], [ 97,  -2], [101,   1], [102,   1], [103,   2]])
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
plt.plot(X[:,0], X[:,1], 'ro')
plt.show()
```

```python
# getting the first and second principal components and creating the plot with new axes
pca_2 = PCA(n_components=2)
pca_2.fit(X)
print(pca_2)
print("Information explained:", pca_2.explained_variance_ratio_)
print("Total Information explained:", sum(pca_2.explained_variance_ratio_)*100, "%")
X_trans_2 = pca_2.transform(X)
print(X_trans_2)
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
ax.scatter(X_trans_2[:, 0], X_trans_2[:, 1], color='black')
for xi, yi in zip(X_trans_2[:, 0], X_trans_2[:, 1]):
    ax.plot([xi, xi], [0, yi], color='red', linestyle='--', linewidth=1)
    ax.plot([0, xi], [yi, yi], color='blue', linestyle='--', linewidth=1)
ax.axhline(0, color='black', linestyle='-', linewidth=1)
ax.axvline(0, color='black', linestyle='-', linewidth=1)
ax.set_xlabel('Normalized new x axis')
ax.set_ylabel('Normalized new y axis')
ax.set_title('First and Second Principal Components')
plt.ylim(-2, 2)
plt.show()
```

```python
# transforming the data back into the previous axes, No information loss occured
X_reduced_2 = pca_2.inverse_transform(X_trans_2)
print(X_reduced_2)
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
plt.plot(X_reduced_2[:,0], X_reduced_2[:,1], 'ro')
plt.show()
```

```python
# getting the first principal component and creating the plot with new axis
pca_1 = PCA(n_components=1)
pca_1.fit(X)
print(pca_1)
print("Information explained:", pca_1.explained_variance_ratio_)
print("Total Information explained:", sum(pca_1.explained_variance_ratio_)*100, "%")
X_trans_1 = pca_1.transform(X)
print(X_trans_1)
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
ax.scatter(X_trans_2[:, 0], X_trans_2[:, 1], color='black')
for xi, yi in zip(X_trans_2[:, 0], X_trans_2[:, 1]):
    ax.plot([xi, xi], [0, yi], color='red', linestyle='--', linewidth=1)
ax.axhline(0, color='black', linestyle='-', linewidth=1)
ax.set_xlabel('Normalized new x axis')
ax.set_ylabel('Normalized new y axis')
ax.set_title('First Principal Component')
plt.ylim(-2, 2)
plt.show()
```

```python
# transforming the data back into the previous axes, Information loss occured
X_reduced_1 = pca_1.inverse_transform(X_trans_1)
print(X_reduced_1)
fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))
ax.plot(X_reduced_1[:,0], X_reduced_1[:,1], 'ro')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('First Principal Component')
plt.show()
```
