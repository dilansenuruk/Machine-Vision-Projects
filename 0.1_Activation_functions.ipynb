# Activation Functions

Activation functions are mathematical functions used in neural networks to introduce non-linearity into the model. Without activation functions, a neural network would behave like a linear model, regardless of its depth, and would not be able to learn complex patterns.

Given a neuron with input vector $\mathbf{x}$, weights $\mathbf{w}$, and bias $b$, the output of the neuron before activation is:

$$
z = \mathbf{w}^T \mathbf{x} + b
$$

The activation function $f(\cdot)$ is then applied to produce the final output:

$$
a = f(z)
$$

Different activation functions are used depending on the problem.  
Activation functions play a crucial role in enabling neural networks to approximate complex, non-linear relationships in data.

```python
# Importing essential libraries
import numpy as np
import matplotlib.pyplot as plt
```

### Linear Activation Function

$$
f(x) = x
$$

```python
x = np.linspace(-10, 10, 100)

def linear(x):
    return x

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, linear(x))
plt.title("Linear Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Sigmoid Activation Function

$$
f(x) = \frac{1}{1 + e^{-x}}
$$
Maps values to the range $(0,1)$, useful for binary classification.

```python
x = np.linspace(-10, 10, 100)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### ReLU Activation Function

$$
f(x) = \max(0, x)
$$
  Widely used due to simplicity and efficiency.

```python
x = np.linspace(-10, 10, 100)

def relu(x):
    return np.maximum(0, x)

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, relu(x))
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Tanh Activation Function

$$
f(x) = \tanh(x)
$$
  Outputs values in the range $(-1,1)$ and is zero-centered.

```python
x = np.linspace(-10, 10, 100)

def tanh(x):
    return np.tanh(x)

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, tanh(x))
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Leaky ReLU Activation Function

$$
f(x) =
\begin{cases}
x & x > 0 \\
\alpha x & x \leq 0
\end{cases}
$$
  Helps to avoid the dying ReLU problem.

```python
x = np.linspace(-10, 10, 100)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Swish Activation Function

$$
f(x) = x \cdot \frac{1}{1 + e^{-x}}
$$
  A smooth, non-monotonic function that often performs better in deep networks.

```python
x = np.linspace(-10, 10, 100)

def swish(x):
    return x * (1 / (1 + np.exp(-x)))

plt.figure(figsize=(5,3))
plt.axhline(y=0, color="red", linewidth=0.5)
plt.axvline(x=0, color="red", linewidth=0.5)
plt.plot(x, swish(x))
plt.title("Swish Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Activation Functions Comparison

```python
x = np.linspace(-10, 10, 200)

plt.figure(figsize=(8, 5))
plt.plot(x, linear(x), label="Linear")
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.plot(x, swish(x), label="Swish")

plt.xlim(-5, 5)
plt.ylim(-1.5, 1.5)
plt.title("Activation Functions Comparison")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()
```
